### [While loop]

- C와의 차이점

Condition에 () 사용하지 않음

Condition 끝에 ‘:’ 반드시 필요

Action 문들의 intentation 맞춤 중요

````python
i=1
sum=0
while i <= 100:
	sum += i # sum = sum + i
	i += 1 # i = i + 1
print(f"Adding from 1 to 100 yields {sum}")

# Adding from 1 to 100 yields 5050
````

- 짝수만 더하기

If 문을 이용

````python
i = 1
sum = 0
while i <= 100:
    if i % 2 == 0:
        sum += i
    i += 1

print(f"{sum}")

# 2550
````

- 홀수는 무시하기 (짝수만 더하기)

if문과 Continue 이용해보기

````python
i=1
sum=0
while i <= 100:
    if i%2 == 1:
        i += 1
        continue
    sum +=i
    i += 1

print (f"Adding from 1 to 100 yields {sum}")
````

> enumerate 함수를 이용해 순서와 값 동시에 가져오기

-> enurmerate 함수로 순서, 값 함께 가져올 수 있음.

````python
marks=[90, 25, 67, 45, 80]

for i, mark in enumerate(marks):

    if mark >= 60:
        print (f"{i}번 학생은 합격입니다")
    else:
        print (f"{i}번 학생은 불합격입니다")
````

>> 기존 형식

````python
marks=[90, 25, 67, 45, 80]

number=0
for mark in marks:
    if mark >= 60:
        print (f"{number}번 학생은 합격입니다")
    else:
        print (f"{number}번 학생은 불합격입니다")
    number +=1
````

> indexing을 활용한 방식

````python
marks=[90, 25, 67, 45, 80]

number=0
for i in range(len(marks)):
    if marks[i] >= 60:
        print (f"{i}번 학생은 합격입니다")
    else:
        print (f"{i}번 학생은 불합격입니다")
````

> for과 while 비교

일정 범위의 것을 대상으로 반복적인 operation을 적용하는 경우, for while보다 코드가 단순

범위가 정해지지 않고, 특정 조건을 만족하는 경우에만 operation을 반복해야 하는 경우, for를 사용해서 구현할 수 없고, while만 가능

Image

````python
# while
list1 = [13, 21, 12, 14, 30, 18]

i = 0
while i < len(list1):
    item = list1[i]
    i += 1
    if not (item % 3) and item < 20:
        print(item, end=" ")
````

````python
# for
list1 = [13, 21, 12, 14, 30, 18]

for i in list1:
    if not (i % 3) and i < 20:
        print(i, end=" ")
````
