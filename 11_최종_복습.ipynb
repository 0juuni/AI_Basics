```python
# 문제 12

data = [
    ["Michael", "010-1234-5678", "100", "95"],
    ["James", "010-8765-4321", "50", "75"],
    ["Benjamin", "010-4444-5555", "75", "95"],
    ["David", "010-6666-7777", "80", "65"],
    ["Aiden", "010-3333-2222", "50", "25"]]
columns = ["Name", "Phone", "Mid-term", "Final"]

df = pd.DataFrame(data, columns=columns)

cols_str = ['Name', 'Phone']
cols_numeric=[col for col in columns if col not in cols_str]
df[cols_numeric]=df[cols_numeric].astype(int)
print(df)

# Mean 칼럼 생성

df['Mean']=df[cols_numeric].mean(axis=1)
print(df)

#Grade 입력하기
# 평균 점수를 기준으로 Grade Column 새로 만들기
# 90 점 이상 A
# 80 점 이상 B
# 60 점 이상 C
# 나머지 F

# Grade 컬럼 생성 (for문 사용)
grade_list = []

# Grade 컬럼 생성
for mean in df['Mean']:
    if mean >= 90:
        grade_list.append("A")
    elif mean >= 80:
        grade_list.append("B")
    elif mean >= 60:
        grade_list.append("C")
    else:
        grade_list.append("F")

df['Grade'] = grade_list
print(df)
```
