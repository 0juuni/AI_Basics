
> String indexing

````python
s = "Hello World"
s[0] #첫 번째 글자
# 'H'

s[-1] # 끝에서 첫 번째 글자
# 'd'

s[6:11]
# 'World'

s[6:11:2] #Index 6부터 10까지 한 글자씩 걸러서
# 'Wrd'

s[:5]
# 'Hello'

s[6:] # Index 6부터 마지막 글자까지
# 'World'

s[-4:] # 맨 끝 네 글자
# 'orld'
````

> String formatting

>> (1) % 라는 operator를 사용해서 literal과 variable의 조합으로 새로운 string 생성

Image

## Formatting

1. %s: 문자 (string)
2. %d: 정수 (decimal, integer)
3. %f: 실수 (float)

````python
s1 = "Total cost: $%7.2f" % (356.089777)
s2 = "Total students: %3d, Boys: %2d" % (230, 120)
s3 = s1 + ", " + s2
print(s3)
#Total cost : $ 356.09, Total students: 230, Boys: 120
````

````python
score = 95; name="수험생"
str_score="*%5s의 점수는 %d점입니다. %-5s 화이팅" % (name, score, name)
print(str_score)
# * 수험생의 점수는 95점입니다. 수험생 화이팅
````

>>(2) string의 format() method 사용

Image

````python
a=2; b=3; str_num='구구단 {} x {} = {}' .format(a, b, a *b)
print(str_num)
# 구구단 2 x 3 = 6

a=2; str_num2='제곱 {0} x {0} = {1}' .format(a, a*a)
print(str_num2)
# 제곱 2 x 2 = 4
````

>>(3) f-string

여는 따옴표나 삼중 따옴표 앞에 f 또는 F로 string을 시작

String 내에서 {와 } 사이에 variable이나 literal을 적을 수 있음

Image

````python
score = 0.85
caption = f"This image is {100*(1-score):.2f}% cat and {100*score:.2f}% dog"
print(caption)
# This image is 15.00% cat and 85.00% dog
````

````python
x=1; y=2
expr1=f"{x:2d} + {y:2d} = {x+y:2d}"
expr2=f"{x} + {y} = {x+y}"
display=expr1 + "; " + expr2
print(display)
# 1 + 2=3; 1 + 2 = 3
````

> String methods

```
count(’n‘) : String 내에 ‘n’이 몇 번 나타나는지 세기

find(’a’) : String 내 ‘a’의 첫번째 등장 위치

upper() : String내 글자를 모두 대문자로 바꾸기 (lower()는 소문자 변환)

replace(’ran’, ‘hee’) : String내 모든 ‘ran’을 ‘hee’로 변경
```

Image

````python
print(myFullname)
print(myFullname.count('n'))
print(myFullname.count('an'))

# String 내 'a'의 첫 번째 등장 위치
print(myFullname.find('a'))

# String 내 글자를 모두 대문자로 바꾸기 (lower()는 소문자 변환)
capitalName = myFullname.upper()
print(capitalName)

# String 내 모든 'ran'을 'hee'로 변경
mySisterName = myFullname.replace('ran', 'hee')
print(mySisterName)
````

````python
# 원본 문자열 생성
original_string = "Hello, Wolrd!"

index_to_change = 7 # 변경하고자 하는 위치
new_character = 'F' # 바꾸고자 하는 target 글자

# 문자열 slicing을 사용하여 변경된 문자열 생성
modified_string = original_string[:index_to_change] + new_character + original_string[index_to_change + 1:]

# 결과 출력
print(modified_string)
````
