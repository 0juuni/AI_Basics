````python
# List 내 특정 위치에 있는 item을 변경
L = [1, 2, 3, 4]
print(L[0])  # 첫 번째 item 출력
print(L[-1])  # 마지막 item 출력
print("---------------------------")

# List 내 item 변경
L = [1, 'hello', 2, 'good morning', 3]
L[3] = 'korea'  # 인덱스 3의 item을 'korea'로 변경
print(L)
print("---------------------------")

# List의 '+' 연산은 두 list를 순차적으로 연결
L1 = [1, 3, 5, 7]
L2 = [2, 4, 6, 8]
L = L1 + L2  # L1과 L2를 이어 붙임
print(L)
print("---------------------------")

# List의 끝에 item 추가
L.append('good bye')  # 끝에 'good bye' 추가
print(L)
print("---------------------------")

# List 내 item들을 크기 순서로 재배치 (item들의 data type이 동일한 경우만 가능)
L = [1, 3, 5, 7, 2, 4, 6, 8]
L.sort()  # 오름차순으로 정렬
print(L)
````

> List Methods

Lists are objects, so they have a lot of methods available on them.

- lst.append(item):  adds item to the end of list lst
- lst.extend(item):  adds a sequence item to the list lst
- lst.count(item):   returns the number of occurences of item in list lst
- lst.index(item):   returns the index of the first occurence of item in list lst
- lst.insert(index, item):  inserts item into list just before index
- lst.pop():         removes the item in the list
- lst.pop(i):        removes an item by index and returns the value
- lst.remove(item):  removes first occurrence of item in the list
- lst.reverse():     reverses the order of items in the list
- lst.sort():          sorts the list in-place

## [Set]
“집합”을 표현하기 위함.

→ Element들을 중복되지 않게 관리하는 것이 핵심

→ 합집합, 교집합, 부분집합 등 집합 operation을 위해 설계됨

중괄호{} 사이에 위치하는 element 간에는 순서가 없음

Element들 간의 순서가 의미를 갖지 않으므로 indexing을 사용할 수 없음

> Set operations

* 생성
```python
s={1,23}, s=set([1,2,3]), empty_set=set(), s=set(”123”)
```

- element 추가
```python
s.add(element) : 새로운 element를 s에 추가
```
- element 삭제
```python
s.remove(element) : 주어진 element를 s에서 삭제. 존재하지 않을 경우, KeyError
s.discard(element) : 주어진 element를 s에서 삭제. 존재하지 않아도 error 없음
s.pop() : set s에서 임의의 element를 제거하고 반환 (s가 비어있을 경우, KeyError)
```
- 부분집합과 상위집합 확인
```python
s.issubset(other_set) : s가 other_set의 부분집합인지 확인
s.issuperset(other_set) : s가 other_set을 포함하는 상위집합인지 확인
```

- Set의 크기 확인
```python
len(s) : s의 element 개수 반환
```

- Set의 복사
```python
s.copy() : s를 복사하여 새로운 set을 생성
```

- 집합 연산
```python
합칩합 : s.union(other_set), set | other_set
교집합 : s.intersection(ohter_set), s & other_set
차집합 : s.difference(other_set), s - other_set
대칭차집합 : s.symmetric_difference(other_set), s^other_set
```

- 집합 내 element 전재 여부 확인
```python
in, not in : 결과값은 True(=1) 혹은 False(=0)
```
- 집합 내 모든 element 접근
```python
for element in s:…… : s 내의 element들에 대해 operation 적용 가능
```

## [Dictionary]

ex) 전화번호부의 이름과 전화번호의 쌍

> List를 dictionary로 변환

Image

zip 사용, (key, value) 형태 리스트-> dict로 묶기, (key, value) 형태 set을 dict로 묶기, (key, value) 형태 tuple을 dict로 묶기

````python
# 방법 1: 두 개의 리스트를 dictionary로 변환 -> zip
keys_list = ['a', 'b', 'c']
values_list = [1, 2, 3]
dict_from_lists = dict(zip(keys_list, values_list))
print(dict_from_lists)  # {'a': 1, 'b': 2, 'c': 3}

# 방법 2: (key, value) 형태의 리스트를 dictionary로 변환
items_list = [('a', 1), ('b', 2), ('c', 3)]
dict_from_list = dict(items_list)
print(dict_from_list)  # {'a': 1, 'b': 2, 'c': 3}

# 방법 3: (key, value) 형태의 set을 dictionary로 변환
items_set = {('a', 1), ('b', 2), ('c', 3)}
dict_from_set = dict(items_set)
print(dict_from_set)  # {'a': 1, 'b': 2, 'c': 3}

# 방법 4: (key, value) 형태의 tuple을 dictionary로 변환
items_tuple = (('a', 1), ('b', 2), ('c', 3))
dict_from_tuple = dict(items_tuple)
print(dict_from_tuple)  # {'a': 1, 'b': 2, 'c': 3}
````
