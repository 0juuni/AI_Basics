2) 하나의 리스트를 입력받아 짝수만 화면에 출력하는 print_even()을 작성하시오.

실행 사례)
print_even([1, 3, 2, 10, 12, 11, 15])  --> 2 10 12

````python
def print_even(x):
  for v in x:
    if v % 2 == 0:
      print(v)

print_even([1, 3, 2, 10, 12, 11, 15])
````

2) 사용자가 지정한 text file 내에서 사용자가 지정한 단어가 몇 번 등장하는지 찾는 코드를 작성하시오. 사용자가 지정한 file이 존재하지 않는 경우에는 오류 메시지를 출력하시오.

힌트: input(), open(), count(), try/except 활용하기!

````python
def create_and_count_word_in_file():
    # 파일 이름 입력받기
    file_name = input("파일 이름을 입력하세요: ")

    # 파일이 없는 경우, 생성하고 내용을 입력받기
    try:
        # 파일이 없으면 새로 생성
        with open(file_name, 'r', encoding='utf-8') as file:
            print(f"파일 '{file_name}'이(가) 이미 존재합니다.")
    except FileNotFoundError:
        print(f"파일 '{file_name}'이(가) 존재하지 않습니다. 파일을 새로 생성합니다.")
        with open(file_name, 'w', encoding='utf-8') as file:
            content = input("파일에 추가할 내용을 입력하세요: ")
            file.write(content)
            print(f"내용이 파일 '{file_name}'에 저장되었습니다.")

    # 검색할 단어 입력받기
    search_word = input("찾고자 하는 단어를 입력하세요: ")

    # 단어 등장 횟수 계산
    try:
        with open(file_name, 'r', encoding='utf-8') as file:
            content = file.read()
            word_count = content.count(search_word)
            print(f"'{search_word}' 단어는 파일 '{file_name}'에서 {word_count}번 등장합니다.")
    except FileNotFoundError:
        print(f"파일 '{file_name}'을(를) 열 수 없습니다.")

# 함수 실행
create_and_count_word_in_file()
````
